```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Values for lambda
lambda_values = [1, 4, 10]
colors = ['blue', 'green', 'red']

# Setting up the figure and axis
plt.figure(figsize=(8, 6))

# X values: the number of events (k)
x = np.arange(0, 20)

# Plotting the Poisson distribution for each lambda
for i, lambda_value in enumerate(lambda_values):
    pmf = poisson.pmf(x, lambda_value)
    plt.plot(x, pmf, color=colors[i], marker='o', label=f' = {lambda_value}')

# Adding titles and labels
plt.title('Poisson Distribution for Different  Values', fontsize=16)
plt.xlabel('Number of Events (k)', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
